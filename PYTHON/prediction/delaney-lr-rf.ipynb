```python
import pandas as pd

df = pd.read_csv('delaney_solubility_with_descriptors.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

### Separate data X and y

```python
y = df['logS']
y
```

```python
X = df.drop('logS', axis=1)
X
```

### Data splitting

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

## Model building

### Linear regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

### Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 score (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 score (Train): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression',lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2 Score', 'Test MSE', 'Test R2 Score']
lr_results
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)

rf.fit(X_train, y_train)
```

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
print('RF MSE (Train): ', rf_train_mse)
print('RF R2 score (Train): ', rf_train_r2)
print('RF MSE (Test): ', rf_test_mse)
print('RF R2 score (Train): ', rf_test_r2)
```

```python
rf_results = pd.DataFrame(['Random Forest',rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2 Score', 'Test MSE', 'Test R2 Score']
rf_results
```

#### Model Comparison

```python
df_models = pd.concat([lr_results, rf_results], axis=0).reset_index(drop=True)
df_models
# df_models.reset_index(drop=True)
```

- A good model will have an MSE value closer to zero, indicating a better goodness of fit to the data - LR better than RF
- The lower MSE for LR indicates a better fit on the training data.
- The higher R2 score for LR indicates it explains more variance in the training data.

- LR has both lower MSE and higher R2 scores on the test set, indicating it is the better model compared to RF for your dataset.

### Visualization

```python
import matplotlib.pyplot as plt
import numpy as np

# Convert to numpy arrays
y_train = np.array(y_train)
y_lr_train_pred = np.array(y_lr_train_pred)

plt.figure(figsize=(5, 5))
plt.scatter(x=y_train, y=y_lr_train_pred, c='green', alpha=0.5)

# Fit a polynomial of degree 1
z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)

# Plot the polynomial fit line
plt.plot(y_train, p(y_train), c='red')
plt.ylabel('Predicted LogS')
plt.xlabel('Experimental LogS')

plt.show()
```
