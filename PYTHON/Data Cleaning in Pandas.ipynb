```python
import pandas as pd
import re

def clean_phone_number(phone):
    # Convert phone number to string if itâ€™s not already
    phone = str(phone)
    
    if pd.isna(phone) or phone.strip().lower() == 'n/a':
        return None
    
    # Replace all non-numeric characters with nothing
    phone = re.sub(r'[^\d]', '', phone)
    
    # Check if the cleaned phone number has exactly 10 digits
    if len(phone) == 10:
        # Format it as 123-456-7890
        phone = re.sub(r'(\d{3})(\d{3})(\d{4})', r'\1-\2-\3', phone)
        return phone
    else:
        return None  # Return None if the number doesn't have exactly 10 digits

# Apply the cleaning function
df['Phone_Number'] = df['Phone_Number'].apply(clean_phone_number)
```
